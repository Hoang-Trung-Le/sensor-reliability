```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample reliability DataFrame (replace this with your actual DataFrame)
reliability_data = {
    'Time': ['2023-10-27 10:00', '2023-10-27 10:02', '2023-10-27 10:04'],
    'Sensor1': [0.8, 0.6, 0.7],
    'Sensor2': [0.9, 0.75, 0.65],
    'Sensor3': [0.7, 0.8, 0.85],
    'Sensor4': [0.65, 0.7, 0.9]
}
reliability_df = pd.DataFrame(reliability_data)
reliability_df['Time'] = pd.to_datetime(reliability_df['Time'])

# Find the sensor with the highest reliability at each time step
highest_reliability_sensors = reliability_df[['Sensor1', 'Sensor2', 'Sensor3', 'Sensor4']].idxmax(axis=1)

# Initialize variables to store start and end timestamps for each period
start_time = None
current_sensor = None
gantt_data = []

# Iterate over each row to identify periods of highest reliability for each sensor
for index, row in reliability_df.iterrows():
    if current_sensor is None:
        current_sensor = highest_reliability_sensors[index]
        start_time = row['Time']
    elif highest_reliability_sensors[index] != current_sensor:
        gantt_data.append((start_time, row['Time'], current_sensor))
        current_sensor = highest_reliability_sensors[index]
        start_time = row['Time']

# Append the last period
gantt_data.append((start_time, reliability_df.iloc[-1]['Time'], current_sensor))

# Plot the Gantt chart
fig, ax = plt.subplots()
for start, end, sensor in gantt_data:
    ax.barh(sensor, width=(end - start).total_seconds() / 3600, left=start, height=0.5)

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Sensor')
ax.set_title('Sensor Reliability Gantt Chart')

# Show plot
plt.show()
```

```python
import numpy as np
from config import *

def convert_to_unix(date_str, format="%d-%m-%Y %H:%M:%S"):
    """
    Convert a datetime string to a Unix timestamp using the specified format.

    Args:
        date_str (str): Datetime string.
        format (str): Format of the datetime string (default is '%d-%m-%Y %H:%M:%S').

    Returns:
        int: Unix timestamp.
    """
    try:
        formatted_time = datetime.strptime(date_str, format)
    except ValueError:
        format_no_time = "%d-%m-%Y"
        formatted_time = datetime.strptime(date_str, format_no_time)

    unix_time = int(time.mktime(formatted_time.timetuple()))
    return unix_time



start_time_feat_norm = "15-03-2024 17:30:00"
end_time_feat_norm = "15-03-2024 18:00:00"
feat_mat_voltage_normal = []

for key, value in resampled_data.items():
    feat_mat_voltage_normal = value[parameter_headers][
        (value[time_header] > convert_to_unix(start_time_feat_norm))
        & (value[time_header] < convert_to_unix(end_time_feat_norm))
    ]
    print(feat_mat_voltage_normal)
feat_mat_voltage_normal_np = np.array(feat_mat_voltage_normal)
averages = np.mean(feat_mat_voltage_normal_np, axis=0)
print(averages)
print(type(averages))
    
feat_mat_voltage_fault = pd.read_csv('average_fault_data.csv')
feat_mat_voltage_fault = feat_mat_voltage_fault.iloc[:, 1:]
print(feat_mat_voltage_fault)
faults = feat_mat_voltage_fault.to_numpy()
print(faults)
print(type(faults))
feat_mat_voltage = np.vstack(
    [
        averages,
        faults
    ]
)
print("Feature matrix: ", feat_mat_voltage)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrames for sensors (you will have your own actual data)
sampling_matrix = {
    'Sensor1': pd.DataFrame({
        'Datetime': pd.date_range('2024-01-01', periods=5, freq='D'),
        'Parameter1': [1, 2, 3, 4, 5],
        'Parameter2': [5, 4, 3, 2, 1]
    }).set_index('Datetime'),
    'Sensor2': pd.DataFrame({
        'Datetime': pd.date_range('2024-01-01', periods=5, freq='D'),
        'Parameter1': [2, 3, 4, 5, 6],
        'Parameter2': [6, 5, 4, 3, 2]
    }).set_index('Datetime'),
}
print(sampling_matrix)
# Example reliability_comparison DataFrame
reliability_comparison = pd.DataFrame({
    'Datetime': pd.date_range('2024-01-01', periods=5, freq='D'),
    'Highest Reliability': ['Sensor1', 'Sensor2', 'Sensor1', 'Sensor2', 'Sensor1']
}).set_index('Datetime')
print(reliability_comparison)
# Initialize an empty DataFrame to store the plot data
plot_data = pd.DataFrame(index=reliability_comparison.index)

# Iterate through each timestamp
for timestamp, row in reliability_comparison.iterrows():
    sensor = row['Highest Reliability']
    sensor_data = sampling_matrix[sensor].loc[[timestamp]]  # Use double brackets to keep DataFrame format
    plot_data = pd.concat([plot_data, sensor_data])
print(plot_data)
# Transpose the DataFrame so each sensor parameter becomes a column
plot_data = plot_data.transpose()

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
for parameter in plot_data.columns:
    ax.plot(plot_data.index, plot_data[parameter], label=parameter)

ax.set_title('Parameter Readings from Most Reliable Sensors')
ax.set_xlabel('Timestamp')
ax.set_ylabel('Parameter Value')
ax.legend()
plt.show()
```

```python
import numpy as np

# Example probability distribution
probabilities = np.array([0.1, 0.2, 0.3, 0.4])

# Ensure the probabilities sum to 1
probabilities /= probabilities.sum()

# Number of outcomes
num_outcomes = len(probabilities)

# Calculating entropy H(P)
entropy = -np.sum(probabilities * np.log(probabilities))

# Calculating maximum entropy H_max
max_entropy = np.log(num_outcomes)

# Calculating the discount factor alpha
discount_factor = 1 - (entropy / max_entropy)

print("Entropy (H(P)):", entropy)
print("Maximum Entropy (H_max):", max_entropy)
print("Discount Factor (alpha):", discount_factor)

a = np.arange(8).reshape((2,4))
print(a)
max0 = np.max(a, axis=0, keepdims=True)
max1 = np.max(a, axis=1, keepdims=True)
print(max0)
print(max1)
b = 1 - (a / max0)
print(b)
```

```python
## Load the data
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import time

# Read the original CSV file
df = pd.read_csv("../data/aqms/raw/aqms_1141.csv")

# Initialize lists to store the values for the new columns
datetime_utc = []
timestamp = []
site_id = []
HUMID = []
TEMP = []
PM25 = []
PM10 = []

# Iterate over the rows of the DataFrame
for index, row in df.iterrows():
    # Construct datetime_utc
    hour = row["Hour"]
    if hour == 24:
        date_time_obj = datetime.strptime(row["Date"], "%Y-%m-%d") + timedelta(days=1)
    else:
        date_time_str = row["Date"] + " " + str(row["Hour"]) + ":00"
        date_time_obj = datetime.strptime(date_time_str, "%Y-%m-%d %H:%M")
        
    datetime_utc.append(date_time_obj)
    # Construct timestamp
    

    # Append other columns
    site_id.append(row["Site_Id"])
    if row["Parameter.ParameterCode"] == "HUMID":
        HUMID.append(row["Value"])
    elif row["Parameter.ParameterCode"] == "TEMP":
        TEMP.append(row["Value"])
    elif row["Parameter.ParameterCode"] == "PM2.5":
        if row["Value"] < 0:
            PM25.append(-row["Value"])
        else:
            PM25.append(row["Value"])
    elif row["Parameter.ParameterCode"] == "PM10":
        if row["Value"] < 0:
            PM10.append(-row["Value"])
        else:
            PM10.append(row["Value"])

datetime_utc = list(sorted(set(datetime_utc)))
# print(datetime_utc)
for dt in datetime_utc:
    timestamp.append(int(time.mktime(dt.timetuple())))
# timestamp = list(set(timestamp))            
site_id = list(set(site_id)) * len(list(set(datetime_utc)))
print(len(datetime_utc))
print(len(timestamp))
print(len(site_id))
print(len(HUMID))
print(len(TEMP))
print(len(PM25))
print(len(PM10))
# Create a new DataFrame from the lists
new_df = pd.DataFrame({
    "datetime_utc": datetime_utc,
    "timestamp": timestamp,
    "site_id": site_id,
    "HUMID": HUMID,
    "TEMP": TEMP,
    "PM2.5": PM25,
    "PM10": PM10
})
# Set datetime_utc as the index of the DataFrame
new_df.set_index("datetime_utc", inplace=True)
# Fill missing values using linear interpolation
new_df.interpolate(method='linear', inplace=True)
# Columns to round up
columns_to_round_up = ["HUMID","TEMP","PM2.5","PM10"]
# Round up the values in the specified columns
new_df[columns_to_round_up] = np.round(new_df[columns_to_round_up], 3)
# Write the new DataFrame to a new CSV file
new_df.to_csv("../data/aqms/processed/aqms_1141_processed.csv", index=True, header=True)
```
